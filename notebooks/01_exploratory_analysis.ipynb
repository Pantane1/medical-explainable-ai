# 01 — Exploratory Data Analysis
Understand the cardiovascular dataset before modelling.

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.utils.data_loader import load_sample_data

%matplotlib inline
sns.set_theme(style='darkgrid')
```

```python
(X_train, X_test, y_train, y_test), feature_names = load_sample_data()

df = pd.DataFrame(
    np.vstack([X_train, X_test]),
    columns=feature_names
)
df['target'] = np.concatenate([y_train, y_test])
print(f'Dataset: {len(df)} rows, {len(feature_names)} features')
print(f'Class balance — Positive: {df.target.mean():.1%}')
df.head()
```

```python
df.describe().round(2)
```

```python
# Missing values
print('Missing values per feature:')
print(df.isnull().sum())
```

```python
# Distribution of each feature by target class
fig, axes = plt.subplots(3, 3, figsize=(14, 10))
axes = axes.flatten()
for i, col in enumerate(feature_names):
    df.groupby('target')[col].plot.hist(bins=20, alpha=0.6, ax=axes[i], legend=True)
    axes[i].set_title(col)
    axes[i].set_xlabel('')
plt.tight_layout()
plt.savefig('../logs/feature_distributions.png', dpi=120)
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(9, 7))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.savefig('../logs/correlation_matrix.png', dpi=120)
plt.show()
```
